```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np 
import seaborn as sns
```

```python
hour = pd.read_csv('hour.csv')
```

```python
hour
```

```python
sns.countplot(data=hour, y= "season", palette= 'pastel', order= hour['season'].value_counts().index[:], hue= 'weathersit')

#shows 
```

```python
sns.displot(data=hour, x = "season", bins=20, hue="cnt", multiple="stack")

plt.show()
```
